```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
credit_df = pd.read_csv("../data/processed/credit_cleaned.csv")

credit_df.head()
```

```python
credit_df.info()
```

```python
numerical_columns = credit_df.select_dtypes(include=["int64", "float64"]).columns

categorical_columns = credit_df.select_dtypes(include=["object"]).columns
```

```python
print("Numerical Columns")
print(numerical_columns)

print()

print("Categorical Columns")
print(categorical_columns)
```

```python
credit_df["Risk"] = credit_df["Risk"].map({
    "good": 0,
    "bad": 1
})
```

```python
credit_df["Risk"].value_counts()
```

```python
le = LabelEncoder()
```

```python
for col in categorical_columns:
    if col != "Risk":
        credit_df[col] = le.fit_transform(credit_df[col])
```

```python
credit_df.head()
```

```python
credit_df.info()
```

```python
X = credit_df.drop(columns="Risk")

X.head()
```

```python
y = credit_df["Risk"]

y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
print("Training Features :", X_train.shape)
print("Testing Features  :", X_test.shape)

print()

print("Training Target :", y_train.shape)
print("Testing Target  :", y_test.shape)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```
