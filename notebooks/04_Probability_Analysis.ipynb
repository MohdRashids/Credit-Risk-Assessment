## Conditional Probability

### Business Question

What is the probability that a customer is classified as bad risk **given that** the loan duration is greater than 24 months?

### Formula

P(Bad Risk | Duration > 24)

### Business Insight

This conditional probability helps the bank understand whether longer loan durations are associated with increased credit risk.

```python
long_duration = credit_df[credit_df["Duration"] > 24]
```

```python
len(long_duration)
```

```python
bad_long_duration = long_duration[long_duration["Risk"] == "bad"]
```

```python
len(bad_long_duration)
```

```python
P_bad_given_long = len(bad_long_duration) / len(long_duration)

print(f"P(Bad Risk | Duration > 24 months) = {P_bad_given_long:.3f}")
```

## Conditional Probability: Loan Duration

### Business Question

What is the probability that a customer is classified as bad risk given that the loan duration is greater than 24 months?

### Calculation

- Customers with duration > 24 months = **230**
- Bad-risk customers among them = **102**

P(Bad Risk | Duration > 24 months) = 102 / 230 = **0.443**

### Business Insight

The overall probability of bad risk is **30%**, but it increases to **44.3%** for customers with loan durations greater than 24 months.

This suggests that longer loan durations are associated with higher credit risk and should be considered during credit evaluation.

# Probability Analysis Summary

The conditional probability analysis shows that the probability of bad credit risk increases under certain customer characteristics.

Among the evaluated conditions:

- Duration greater than 24 months has the highest probability of bad risk (44.3%).
- Credit amount greater than 5000 also has a high probability of bad risk (41.5%).
- Customers living in rented housing have a probability of 39.1%.
- Customers younger than 30 years have a probability of 36.9%.
- Male customers have a probability of 27.7%, which is slightly lower than the overall dataset probability.

These findings suggest that loan characteristics such as duration and credit amount are stronger indicators of credit risk than demographic characteristics such as gender.
