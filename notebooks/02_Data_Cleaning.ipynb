 Data Cleaning
The cleaning process includes:
- Removing unnecessary columns
- Handling missing values
- Checking duplicate records
- Validating data types
- Preparing the dataset for Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../data/raw/german_credit_data.csv")

df.head()
```

```python
credit_df = df.copy()
```

```python
credit_df.shape
```

```python
credit_df.columns
```

```python
credit_df.drop(columns="Unnamed: 0", inplace=True)
```

```python
credit_df.head()
```

```python
credit_df.isnull().sum()
```

```python
credit_df["Saving accounts"].value_counts(dropna=False)
```

```python
credit_df["Checking account"].value_counts(dropna=False)
```

```python
credit_df["Saving accounts"] = credit_df["Saving accounts"].fillna("Unknown")
credit_df["Checking account"] = credit_df["Checking account"].fillna("Unknown")
```

```python
credit_df.isnull().sum()
```

```python
credit_df.to_csv("../data/processed/credit_cleaned.csv", index=False)
```
