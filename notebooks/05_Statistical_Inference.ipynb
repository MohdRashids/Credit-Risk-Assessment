```python
import pandas as pd
import numpy as np

from scipy.stats import ttest_ind
from scipy.stats import chi2_contingency

credit_df = pd.read_csv("../data/processed/credit_cleaned.csv")

credit_df.head()
```

```python
good_duration = credit_df[credit_df["Risk"]=="good"]["Duration"]

bad_duration = credit_df[credit_df["Risk"]=="bad"]["Duration"]
```

```python
print("Good Risk Mean :", good_duration.mean())
print("Bad Risk Mean :", bad_duration.mean())
```

```python
t_stat, p_value = ttest_ind(
    good_duration,
    bad_duration,
    equal_var=False
)

print("T Statistic :", t_stat)
print("P Value :", p_value)
```

# Independent t-Test: Credit Amount vs Risk

```python
good_credit = credit_df[credit_df["Risk"] == "good"]["Credit amount"]

bad_credit = credit_df[credit_df["Risk"] == "bad"]["Credit amount"]
```

```python
print("Good Risk Mean:", good_credit.mean())
print("Bad Risk Mean:", bad_credit.mean())
```

```python
t_stat, p_value = ttest_ind(
    good_credit,
    bad_credit,
    equal_var=False
)

print("T Statistic:", t_stat)
print("P Value:", p_value)
```

First Chi-Square Test: Sex vs Risk

```python
sex_table = pd.crosstab(
    credit_df["Sex"],
    credit_df["Risk"]
)

sex_table
```

```python
chi2, p_value, dof, expected = chi2_contingency(sex_table)

print("Chi-Square Statistic:", chi2)
print("P Value:", p_value)
print("Degrees of Freedom:", dof)
```

```python
expected_df = pd.DataFrame(
    expected,
    index=sex_table.index,
    columns=sex_table.columns
)

expected_df
```

```python
housing_table = pd.crosstab(
    credit_df["Housing"],
    credit_df["Risk"]
)

housing_table
```

```python
chi2, p_value, dof, expected = chi2_contingency(housing_table)

print("Chi-Square Statistic:", chi2)
print("P Value:", p_value)
print("Degrees of Freedom:", dof)
```

```python
expected_df = pd.DataFrame(
    expected,
    index=housing_table.index,
    columns=housing_table.columns
)

expected_df
```

```python
purpose_table = pd.crosstab(
    credit_df["Purpose"],
    credit_df["Risk"]
)

purpose_table
```

```python
chi2, p_value, dof, expected = chi2_contingency(purpose_table)

print("Chi-Square Statistic:", chi2)
print("P Value:", p_value)
print("Degrees of Freedom:", dof)
```

```python
expected_df = pd.DataFrame(
    expected,
    index=purpose_table.index,
    columns=purpose_table.columns
)

expected_df
```

# Statistical Inference Summary

Statistical hypothesis testing was performed to validate whether the observed relationships in the dataset were statistically significant.

### Independent t-tests

- Loan duration differs significantly between good-risk and bad-risk customers.
- Credit amount differs significantly between good-risk and bad-risk customers.

### Chi-Square Tests

- Housing status is significantly associated with credit risk.
- Customer sex is significantly associated with credit risk.
- Loan purpose is not significantly associated with credit risk.

### Overall Conclusion

The statistical analysis indicates that **loan duration**, **credit amount**, and **housing status** are the strongest indicators of credit risk in this dataset. These features will be prioritized during feature engineering and model building.
