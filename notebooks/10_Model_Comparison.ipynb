```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
comparison = pd.DataFrame({
    "Model":[
        "Logistic Regression",
        "Balanced Logistic",
        "Decision Tree",
        "Random Forest"
    ],

    "Accuracy":[
        0.73,
        0.64,
        0.70,
        0.77
    ],

    "Bad Recall":[
        0.22,
        0.65,
        0.45,
        0.47
    ],

    "Bad Precision":[
        0.65,
        0.43,
        0.50,
        0.67
    ],

    "ROC AUC":[
        0.690,
        None,
        0.629,
        0.771
    ]
})

comparison
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=comparison,
    x="Model",
    y="Accuracy"
)

plt.title("Accuracy Comparison")

plt.xticks(rotation=15)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=comparison,
    x="Model",
    y="Bad Recall"
)

plt.title("Bad Risk Recall")

plt.xticks(rotation=15)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=comparison,
    x="Model",
    y="Bad Precision"
)

plt.title("Bad Risk Precision")

plt.xticks(rotation=15)

plt.show()
```

```python
roc_df = comparison.dropna()

plt.figure(figsize=(8,5))

sns.barplot(
    data=roc_df,
    x="Model",
    y="ROC AUC"
)

plt.title("ROC-AUC Comparison")

plt.xticks(rotation=15)

plt.show()
```

```python
comparison.sort_values(
    by="Accuracy",
    ascending=False
)
```

```python
print("Best Accuracy Model : Random Forest")

print("Best Bad Risk Recall : Balanced Logistic Regression")

print("Recommended Model for Credit Risk : Random Forest")
```
